```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time
```

```python
url = "https://wuzzuf.net/search/jobs?q=data%20analyst&a=spbg"
```

```python
options = Options()
options.add_argument("--start-maximized")

driver = webdriver.Chrome(options=options)

driver.get(url)

time.sleep(5)
```

```python
print("Page title:", driver.title)
print("Current URL:", driver.current_url)
```

```python
from selenium.webdriver.common.by import By

job_cards = driver.find_elements(By.CSS_SELECTOR, "div.css-ghe2tq")

print("Number of job cards found:", len(job_cards))

for i, card in enumerate(job_cards[:3], start=1):
    print(f"\n--- Job Card {i} ---")
    print(card.text[:500])
```

```python
from selenium.webdriver.common.by import By

jobs_basic = []

for card in job_cards:
    try:
        job_title = card.find_element(By.CSS_SELECTOR, "h2 a").text.strip()
    except:
        job_title = None

    try:
        company_name = card.find_element(By.CSS_SELECTOR, "a.css-ipsyv7").text.strip().replace(" -", "")
    except:
        company_name = None

    try:
        location = card.find_element(By.CSS_SELECTOR, "span.css-16x61xq").text.strip()
    except:
        location = None

    try:
        posted_date = card.find_element(By.CSS_SELECTOR, "div.css-1k5ee52 > div").text.strip()
    except:
        posted_date = None

    try:
        tags = [tag.text.strip() for tag in card.find_elements(By.CSS_SELECTOR, "span.eoyjyou0")]
        job_type = tags[0] if len(tags) > 0 else None
        workplace_type = tags[1] if len(tags) > 1 else None
    except:
        job_type = None
        workplace_type = None

    try:
        experience_text = card.find_element(By.CSS_SELECTOR, "div.css-1rhj4yg div:nth-child(2)").text.strip()
    except:
        experience_text = None

    jobs_basic.append({
        "job_title": job_title,
        "company_name": company_name,
        "location": location,
        "posted_date": posted_date,
        "job_type": job_type,
        "workplace_type": workplace_type,
        "experience_text": experience_text
    })

print("Extracted jobs:", len(jobs_basic))

for job in jobs_basic[:3]:
    print(job)
```

```python
job_links = []

for card in job_cards:
    try:
        link_element = card.find_element(By.CSS_SELECTOR, "h2 a")
        job_url = link_element.get_attribute("href")

        if job_url.startswith("/"):
            job_url = "https://wuzzuf.net" + job_url

        job_links.append(job_url)

    except Exception as e:
        print("Error extracting link:", e)

print("Number of job links:", len(job_links))

for link in job_links[:5]:
    print(link)
```

```python
from selenium.webdriver.common.by import By

details = {}

# title
try:
    details["title"] = driver.find_element(By.CSS_SELECTOR, "h1.css-gkdl1m").text.strip()
except:
    details["title"] = None

# company + location text
try:
    details["company_location_text"] = driver.find_element(By.CSS_SELECTOR, "strong.css-1vlp604").text.strip()
except:
    details["company_location_text"] = None

# posted date
try:
    details["posted_date_details"] = driver.find_element(By.CSS_SELECTOR, "span.css-154erwh").text.strip()
except:
    details["posted_date_details"] = None

# job details rows: experience, salary, etc.
details["experience"] = None
details["salary"] = None

rows = driver.find_elements(By.CSS_SELECTOR, "div.css-1ajx53j")

for row in rows:
    row_text = row.text.strip()

    if row_text.startswith("Experience Needed:"):
        details["experience"] = row_text.replace("Experience Needed:", "").strip()

    elif row_text.startswith("Salary:"):
        details["salary"] = row_text.replace("Salary:", "").strip()

# skills
try:
    skill_elements = driver.find_elements(By.CSS_SELECTOR, "div.css-qe7mba span.css-1vi25m1")
    details["skills"] = [skill.text.strip() for skill in skill_elements if skill.text.strip()]
except:
    details["skills"] = []

# job description
try:
    details["job_description"] = driver.find_element(
        By.XPATH,
        "//h2[contains(text(), 'Job Description')]/following-sibling::div"
    ).text.strip()
except:
    details["job_description"] = None

# job requirements
try:
    details["job_requirements"] = driver.find_element(
        By.XPATH,
        "//h2[contains(text(), 'Job Requirements')]/following-sibling::div"
    ).text.strip()
except:
    details["job_requirements"] = None

details
```

```python
import time
from selenium.webdriver.common.by import By
from urllib.parse import quote

search_keywords = [
    # core
    "data analyst",
    "data engineer",
    "data scientist",
    "business analyst",
    "analytics engineer",

    # BI & SQL
    "sql developer",
    "sql analyst",
    "power bi",
    "tableau developer",
    "bi developer",

    # python & ETL
    "python developer data",
    "etl developer",
    "data pipeline engineer",

    # big data
    "big data engineer",
    "data warehouse engineer",

    # AI
    "machine learning engineer",
    "ai engineer",

    # analytics domains
    "financial analyst",
    "operations analyst"
]

all_job_links = []

max_pages_per_keyword = 4

for keyword in search_keywords:
    encoded_keyword = quote(keyword)
    base_url = f"https://wuzzuf.net/search/jobs?q={encoded_keyword}"

    print(f"\n==============================")
    print(f"Keyword: {keyword}")
    print(f"==============================")

    for page in range(max_pages_per_keyword):
        page_url = f"{base_url}&start={page}"

        print(f"Scraping page {page}: {page_url}")

        driver.get(page_url)
        time.sleep(6)

        job_cards = driver.find_elements(By.CSS_SELECTOR, "div.css-ghe2tq")
        print("Cards found:", len(job_cards))

        if len(job_cards) == 0:
            print("No cards found, stopping this keyword.")
            break

        for card in job_cards:
            try:
                link = card.find_element(By.CSS_SELECTOR, "h2 a").get_attribute("href")

                if link.startswith("/"):
                    link = "https://wuzzuf.net" + link

                all_job_links.append({
                    "search_keyword": keyword,
                    "job_url": link
                })

            except Exception as e:
                print("Error extracting link:", e)

        time.sleep(5)

print("\nTotal collected links before duplicates:", len(all_job_links))

unique_links = {}

for item in all_job_links:
    url = item["job_url"]
    keyword = item["search_keyword"]

    if url not in unique_links:
        unique_links[url] = {
            "job_url": url,
            "search_keywords": [keyword]
        }
    else:
        unique_links[url]["search_keywords"].append(keyword)

all_job_links_unique = list(unique_links.values())

print("Total unique links:", len(all_job_links_unique))

for item in all_job_links_unique[:10]:
    print(item)
```

```python
import time
from datetime import datetime
from selenium.webdriver.common.by import By
import pandas as pd
scraped_jobs = []

test_links = all_job_links_unique

for index, item in enumerate(test_links, start=1):
    job_url = item["job_url"]
    search_keywords = item["search_keywords"]

    print(f"\nScraping job {index}/{len(test_links)}")
    print(job_url)

    driver.get(job_url)
    time.sleep(5)

    job = {
        "job_url": job_url,
        "source_platform": "Wuzzuf",
        "search_keywords": search_keywords,
        "scraped_at": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    }

    try:
        job["job_title"] = driver.find_element(By.CSS_SELECTOR, "h1.css-gkdl1m").text.strip()
    except:
        job["job_title"] = None

    try:
        company_location_text = driver.find_element(By.CSS_SELECTOR, "strong.css-1vlp604").text.strip()
        job["company_location_text"] = company_location_text
    except:
        job["company_location_text"] = None

    try:
        job["posted_date"] = driver.find_element(By.CSS_SELECTOR, "span.css-154erwh").text.strip()
    except:
        job["posted_date"] = None

    try:
        tags = driver.find_elements(By.CSS_SELECTOR, "span.eoyjyou0")
        tag_texts = [tag.text.strip() for tag in tags if tag.text.strip()]
        job["job_type"] = tag_texts[0] if len(tag_texts) > 0 else None
        job["workplace_type"] = tag_texts[1] if len(tag_texts) > 1 else None
    except:
        job["job_type"] = None
        job["workplace_type"] = None

    job["experience"] = None
    job["salary"] = None

    rows = driver.find_elements(By.CSS_SELECTOR, "div.css-1ajx53j")

    for row in rows:
        row_text = row.text.strip()

        if row_text.startswith("Experience Needed:"):
            job["experience"] = row_text.replace("Experience Needed:", "").strip()

        elif row_text.startswith("Salary:"):
            job["salary"] = row_text.replace("Salary:", "").strip()

    try:
        skill_elements = driver.find_elements(By.CSS_SELECTOR, "div.css-qe7mba span.css-1vi25m1")
        job["skills"] = [skill.text.strip() for skill in skill_elements if skill.text.strip()]
    except:
        job["skills"] = []

    try:
        job["job_description"] = driver.find_element(
            By.XPATH,
            "//h2[contains(text(), 'Job Description')]/following-sibling::div"
        ).text.strip()
    except:
        job["job_description"] = None

    try:
        job["job_requirements"] = driver.find_element(
            By.XPATH,
            "//h2[contains(text(), 'Job Requirements')]/following-sibling::div"
        ).text.strip()
    except:
        job["job_requirements"] = None

    scraped_jobs.append(job)

    time.sleep(3)

print("\nScraped jobs:", len(scraped_jobs))



df = pd.DataFrame(scraped_jobs)

df.to_csv("wuzzuf_jobs_raw.csv", index=False, encoding="utf-8-sig")

print("CSV file saved successfully")
```
